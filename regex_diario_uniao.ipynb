```python
import requests 
import bs4 
  
text= "SHOCK INSTALACOES E MANUTENCAO LTDA"
url = 'https://google.com/search?q=' + text 
request_result=requests.get( url ) 
soup = bs4.BeautifulSoup(request_result.text, 
                         "html.parser") 
print(soup)
```

```python

link =soup.find_all("div", attrs={"class": "kCrYT"}) 
for i in link:
    texto = i.find('h3')

    if texto is None:        
        continue
    else:
        texto = texto.getText()
        print(texto)
    
    url = i.find('a', href=re.compile(r'[/]([a-z]|[A-Z])\w+'))
    url = str(url)
    
    url_parte = url.split('url?q=')
    if len(url_parte) >=2:
        url_parte_2 = url_parte[1].split('&amp;sa')
        print(url_parte_2[0])
    
    
    print('------------------------------------')
```
